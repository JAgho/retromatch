```python
def find_blocks(y):
    onsets = librosa.onset.onset_detect(y=y,pre_max=3, post_max=3, pre_avg=3, post_avg=5, delta=0.5, wait=10)
    diffs = np.diff(onsets)
    bigdiffs = np.argwhere(diffs > 50.0)
    bigger_diffs = np.diff(bigdiffs[:,0])
    print(np.shape(bigdiffs))
    return bigger_diffs
    
```
