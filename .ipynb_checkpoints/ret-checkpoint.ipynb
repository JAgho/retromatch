```python
def find_blocks(y):
    onsets = librosa.onset.onset_detect(y=y)
    diffs = np.diff(onsets)
    bigdiffs = diffs > 100.0
    return onsets[bigdiffs]
    
```

```python
find_blocks(y)
```
